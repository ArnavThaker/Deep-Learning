### Problem 1

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision

def model(inputSize, numClasses):
    return nn.Linear(inputSize, numClasses)

inputSize = 784
numClasses = 10
trainCorrect = 0
trainTotal = 0
testCorrect = 0
testTotal = 0

myModel = model(inputSize, numClasses)
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(myModel.parameters(), lr = 0.05)
MNISTTransform = torchvision.transforms.Compose([torchvision.transforms.ToTensor()])

trainSet = torchvision.datasets.MNIST(root = './data', train = True, download = True, transform = MNISTTransform)
trainLoader = torch.utils.data.DataLoader(trainSet, batch_size = 10, shuffle = True)

testSet = torchvision.datasets.MNIST(root = './data', train = False, download = True, transform = MNISTTransform)
testLoader = torch.utils.data.DataLoader(testSet, batch_size = 1, shuffle = False)

numEpochs = 10

for epoch in range(numEpochs):
    for i, (images, labels) in enumerate(trainLoader):
        images = images.reshape(images.size(0), 1*28*28)
        
        outputs = myModel(images)
        loss = criterion(outputs, labels)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        _, predicted = torch.max(outputs, 1)
        trainTotal += labels.size(0)
        trainCorrect += (predicted == labels).sum().item()
        
    for i, (images, labels) in enumerate(testLoader):
        images = images.reshape(images.size(0), 1*28*28)
        outputs = myModel(images)
        _, predicted = torch.max(outputs, 1)
        testTotal += labels.size(0)
        testCorrect += (predicted == labels).sum().item()

    
    print('Epoch: {}, Training Accuracy: {}%, Test Accuracy: {}%' .format(epoch + 1, (trainCorrect / trainTotal) * 100, (testCorrect / testTotal) * 100))

```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision

def model(inputSize, numClasses):
    return nn.Linear(inputSize, numClasses)


    
inputSize = 784
numClasses = 10
trainCorrect = 0
trainTotal = 0
testCorrect = 0
testTotal = 0

myModel = model(inputSize, numClasses)
criterion = nn.CrossEntropyLoss()
optimizerSGD = optim.SGD(myModel.parameters(), lr = 0.1)
optimizerGD = optim.SGD(myModel.parameters(), lr = 0.1)
MNISTTransform = torchvision.transforms.Compose([torchvision.transforms.ToTensor()])

trainSet = torchvision.datasets.MNIST(root = './data', train = True, download = True, transform = MNISTTransform)

testSet = torchvision.datasets.MNIST(root = './data', train = False, download = True, transform = MNISTTransform)


def extraction():
    filters = (trainSet.targets == 0) | (trainSet.targets == 1) | (trainSet.targets == 3)| (trainSet.targets == 4) | (trainSet.targets == 7)
    trainSet.targets = trainSet.targets[filters]
    trainSet.data = trainSet.data[filters]
    
extraction()


trainSGDLoader = torch.utils.data.DataLoader(trainSet, batch_size = 1, shuffle = True)
testSGDLoader = torch.utils.data.DataLoader(testSet, batch_size = 1, shuffle = False)

trainGDLoader = torch.utils.data.DataLoader(trainSet, batch_size = inputSize, shuffle = True)
testGDLoader = torch.utils.data.DataLoader(testSet, batch_size = inputSize, shuffle = True)

numEpochs = 3

for epoch in range(numEpochs):
    for i, (images, labels) in enumerate(trainGDLoader):
        images = images.reshape(images.size(0), 1*28*28)
        
        outputs = myModel(images)
        lossGD = criterion(outputs, labels)
        
        optimizerGD.zero_grad()
        lossGD.backward()
        optimizerGD.step()
        
        _, predicted = torch.max(outputs, 1)
        trainTotal += labels.size(0)
        trainCorrect += (predicted == labels).sum().item()
    GDTrainAccuracy = 100 * trainCorrect / trainTotal
    
    for i, (images, labels) in enumerate(testGDLoader):
        images = images.reshape(images.size(0), 1*28*28)
        outputs = myModel(images)
        
        _, predicted = torch.max(outputs, 1)
        testTotal += labels.size(0)
        testCorrect += (predicted == labels).sum().item()
    GDTestAccuracy = 100 * testCorrect / testTotal
    
    print('Gradient Descent, Epoch: {}, Training Accuracy: {}%, Test Accuracy: {}%' 
          .format(epoch + 1, (trainCorrect / trainTotal) * 100, (testCorrect / testTotal) * 100))

trainCorrect = 0
trainTotal = 0
testCorrect = 0
testTotal = 0

for epoch in range(numEpochs):
    for i, (images, labels) in enumerate(trainSGDLoader):
        images = images.reshape(images.size(0), 1*28*28)
        
        outputs = myModel(images)
        lossSGD = criterion(outputs, labels)
        
        optimizerSGD.zero_grad()
        lossSGD.backward()
        optimizerSGD.step()
        
        _, predicted = torch.max(outputs, 1)
        trainTotal += labels.size(0)
        trainCorrect += (predicted == labels).sum().item()
    SGDTrainAccuracy = 100 * trainCorrect / trainTotal
    
    for i, (images, labels) in enumerate(testSGDLoader):
        images = images.reshape(images.size(0), 1*28*28)
        outputs = myModel(images)
        
        _, predicted = torch.max(outputs, 1)
        testTotal += labels.size(0)
        testCorrect += (predicted == labels).sum().item()
    SGDTestAccuracy = 100 * testCorrect / testTotal
    
    print('Stochastic Gradient Descent, Epoch: {}, Training Accuracy: {}%, Test Accuracy: {}%' 
          .format(epoch + 1, (trainCorrect / trainTotal) * 100, (testCorrect / testTotal) * 100))


```
